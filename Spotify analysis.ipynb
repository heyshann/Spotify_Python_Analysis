```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dftrack=pd.read_csv('tracks.csv')
dftrack.head()
```

```python
# checking null values column wise
pd.isnull(dftrack).sum()
```

```python
dftrack.info()
```

```python
sorted_df = dftrack.sort_values('popularity' , ascending=True)
sorted_df.head(10)
```

```python
dftrack.describe().transpose()
```

```python
mostpop=dftrack.query('popularity>90' , inplace = False).sort_values('popularity' , ascending=False)
mostpop[:10]
```

```python
dftrack.set_index("release_date",inplace= True)
dftrack.index=pd.to_datetime(dftrack.index)
dftrack.head()
```

```python
#to check artist in 18th row
dftrack[['artists']].iloc[18]
```

```python
dftrack['duration']=dftrack['duration_ms'].apply(lambda x:round(x/1000))
dftrack.drop('duration_ms',inplace=True,axis=1)
```

```python
dftrack.duration.head()
```

```python
dftrack
```

```python
corr_df=dftrack.drop(["key","mode","explicit"],axis=1)
corr_df.corr(method="pearson")
```

```python
dataplot = sns.heatmap(corr_df.corr(method="pearson"), cmap="inferno", annot=True,linewidth=1, linecolor="Black",fmt=".1g")
dataplot.set_title("Correlation HeatMap b/w Varible")
```

```python
sample_df=dftrack.sample(int(0.004*len(dftrack)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set_title("Loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="b").set_title("Popularity vs Acousticness")
```

```python
dftrack['dates']=dftrack.index.get_level_values('release_date')
dftrack.dates=pd.to_datetime(dftrack.dates)
years=dftrack.dates.dt.year
```

```python
sns.displot(years, discrete=True, height=5, aspect=2, kind="hist").set(title="No of Songs per Year")
```

```python
total_dr=dftrack.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Years vs Duration")
```

```python
total_dr=dftrack.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Years vs Duration")
```

#### Genre Analysis

```python
dfgenre=pd.read_csv("SpotifyFeatures.csv")
dfgenre.head()
```

```python
dfgenre.shape
```

```python
plt.title("Durations of songs in diff genre")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre",x ="duration_ms", data=dfgenre)
plt.xlabel("duration in milli sec")
plt.ylabel("Genre")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=dfgenre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y="genre", x="popularity",data=famous).set(title="Top 5 genre")
```
